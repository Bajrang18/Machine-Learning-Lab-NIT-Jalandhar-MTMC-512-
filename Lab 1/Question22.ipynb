### 22. Load a dataset or create a dummy one to perform the following tasks.
##### a) Identify missing values using isnull() and notnull().
##### b) Handle missing values using dropna() and fillna() (mean/median)
##### c) Detect and remove duplicate rows.
##### d) Rename columns in a DataFrame.
##### e) Change data types of a column.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'A':[1,2,np.nan,2],
    'B':[5,5,6,6]
})

# a) Missing values
df.isnull()
df.notnull()

# b) Handle missing
df.fillna(df.mean())
df.dropna()

# c) Remove duplicates
df.drop_duplicates()

# d) Rename columns
df.rename(columns={'A':'Alpha','B':'Beta'}, inplace=True)

# e) Change datatype
df['Beta'] = df['Beta'].astype(float)
```
