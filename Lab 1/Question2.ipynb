### 2. Create the following arrays:
#### a) zeros matrix of shape (4,4)
#### b) identity matrix of size 5
#### c) random array of size 10

```python
import numpy as np
zero_arr = np.zeros((4,4))
print("Zero Matrix", zero_arr)
identity_arr = np.eye(5)
print("identity matrix \n", identity_arr)
rand_arr = np.random.random(10)
print("Random Array",rand_arr)
```
