### 18. Generate 1,000 samples of size 𝑛 = 30 from a highly skewed distribution (like an Exponential or Beta distribution). Calculate the mean of each sample. Plot a histogram of these 1,000 means—what shape does it take?

```python
import matplotlib as plt
import numpy as np
means = [np.mean(np.random.exponential(1,30)) for _ in range(1000)]

import matplotlib.pyplot as plt
plt.hist(means, bins=30)
plt.show()
```
