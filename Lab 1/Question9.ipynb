### 9. Given an array of non-negative integers (e.g., [1, 2, 1, 3, 3, 3, 0]), find the most frequent value (mode) using only np.bincount and np.argmax.

```python
import numpy as np
arr = np.array([1,2,1,3,3,3,0])
most_frequent_Element = np.argmax(np.bincount(arr))
print("The Most Frequent Element is",most_frequent_Element)
```
