### 5. Generate a 2D array and flatten it using flatten() and ravel(). Explain the difference.

```python
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
print("Originaly Generated array", arr)
arr1=arr.flatten()
print("Flattened array using flatten()", arr1)
arr2=arr.ravel()
print("Flattened array using ravel()", arr2)
```

Both flatten() and ravel() are used to convert a multi-dimensional array into a 1D (one-dimensional) array. The primary difference lies in memory management: flatten() always returns a copy, whereas ravel() returns a view (reference) of the original array whenever possible.
