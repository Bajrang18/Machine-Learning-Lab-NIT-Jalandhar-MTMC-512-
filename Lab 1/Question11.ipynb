### 11. Create a DataFrame with an index starting from 10 to 15. Demonstrate the difference between .loc[10:12] and .iloc[10:12]. Why does one return data while the other likely throws an error or returns an empty set?

```python
import pandas as pd
data = {'A', 'B', 'C', 'D', 'E', 'F'}
data_frame = pd.DataFrame(data, index=[10,11,12,13,14,15])
print(data_frame.loc[10:12])
data_frame.iloc[10:12]

```

.loc uses index labels

.iloc uses position numbers
