### 19. Given a 2𝐷 array of data, calculate the Covariance Matrix using only NumPy’s matrix multiplication (@) and subtraction—do not use np.cov

```python
import numpy as np
X = np.random.rand(5,3)
X_centered = X - X.mean(axis=0)
cov = (X_centered.T @ X_centered) / (X.shape[0]-1)
print("Covariance matrix:\n", cov)
```
