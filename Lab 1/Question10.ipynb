### 10. You have two arrays representing coordinates: 𝐴 = (𝑛, 2) and 𝐵 = (𝑚, 2). Calculate the Euclidean distance between every pair of points in 𝐴 and 𝐵 without using a single for loop.

```python
import numpy as np
A = np.random.rand(3,2)
B = np.random.rand(4,2)

distances = np.sqrt(((A[:, None, :] - B[None, :, :]) ** 2).sum(axis=2))
print(distances)
```
