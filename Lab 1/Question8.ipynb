### 8. Create a 10 × 10 array of random integers. Find the indices of the maximum value in the array and replace that specific value with −1 without using a loop.

```python
import numpy as np
arr = np.random.randint(1,100,(10,10))
idx = np.unravel_index(np.argmax(arr), arr.shape)
arr[idx] = -1
print(arr)
```
