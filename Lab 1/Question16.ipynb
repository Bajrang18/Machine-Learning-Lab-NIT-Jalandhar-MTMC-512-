### 16. Write a function that takes a Pandas Series, calculates the Z-scores for every element, and returns a list of values that are more than 3 standard deviations from the mean.

```python
import pandas as pd

def find_outliers(series):
    mean = series.mean()
    std = series.std()
    
    z_scores = (series - mean) / std
    outliers = series[abs(z_scores) > 3]
    
    return outliers.tolist()
data = pd.Series([10, 12, 11, 9, 10, 11, 5000])
result = find_outliers(data)
print(result)
```

##### Z-score measures how far a data point is from the mean in terms of standard deviation.
