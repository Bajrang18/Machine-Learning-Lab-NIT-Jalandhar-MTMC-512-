### 14. Create a DataFrame with a daily DatetimeIndex for the year 2025. Generate random stock prices. Calculate the 7-day rolling average and find the specific date with the highest volatility (standard deviation) within a 30-day window.

```python
import numpy as np
import pandas as pd
dates = pd.date_range('2025-01-01','2025-12-31',freq='D')
prices = pd.Series(np.random.rand(len(dates))*100, index=dates)

rolling_avg = prices.rolling(7).mean()
volatility = prices.rolling(30).std()

max_vol_date = volatility.idxmax()
print(max_vol_date)
```
