### 12. Load a dataset (or create a dummy one) with two columns: Category and Sales. If Sales has missing values (NaN), fill them with the median sales value of their specific Category (not the global median).

```python
import pandas as pd
df = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B'],
    'Sales': [100, None, 200, None]
})

df['Sales'] = df.groupby('Category')['Sales'].transform(
    lambda x: x.fillna(x.median())
)
print(df)
```
