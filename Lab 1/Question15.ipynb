### 15. Take a "Wide" format DataFrame (columns: Year, Q1_Profit, Q2_Profit, Q3_Profit, Q4_Profit) and transform it into a "Long" format (columns: Year, Quarter, Profit) using pd.melt.

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'Year':[2023,2024],
    'Q1_Profit':[10,15],
    'Q2_Profit':[12,18],
    'Q3_Profit':[14,20],
    'Q4_Profit':[16,22]
})

long_df = pd.melt(df, id_vars='Year',
                  var_name='Quarter',
                  value_name='Profit')
print(long_df)
```
