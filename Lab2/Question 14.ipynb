## 14. Generate a random dataset having two classes or load an existing dataset. Train: One classifier optimized for log-likelihood, One optimized for accuracy Compare probability confidence and decision behavior.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

np.random.seed(42)

X, y = make_classification(
    n_samples=1000,
    n_features=5,
    n_informative=3,
    class_sep=1.0
)

logreg = LogisticRegression(max_iter=500)
logreg.fit(X, y)

proba_lr = logreg.predict_proba(X)[:, 1]
y_pred_lr = logreg.predict(X)


tree = DecisionTreeClassifier()
tree.fit(X, y)

proba_tree = tree.predict_proba(X)[:, 1]
y_pred_tree = tree.predict(X)



print("Logistic Regression Accuracy:",
      accuracy_score(y, y_pred_lr))

print("Decision Tree Accuracy:",
      accuracy_score(y, y_pred_tree))


print("Logistic Regression probabilities:",
      np.percentile(proba_lr, [5, 50, 95]))

print("Decision Tree probabilities:",
      np.percentile(proba_tree, [5, 50, 95]))
```

### Classifiers optimized for log-likelihood, such as logistic regression, focus on producing accurate probability estimates and penalize overconfident incorrect predictions. In contrast, classifiers optimized for accuracy alone, such as decision trees, prioritize hard classification decisions and often produce extreme and poorly calibrated probabilities. As a result, log-likelihood-optimized models exhibit smoother decision boundaries and more reliable probabilistic confidence, even when accuracy is comparable.
