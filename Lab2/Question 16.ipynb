## 16. Implement GLMs using different exponential family distributions:
### a) Gaussian
### b) Poisson
### c) Binomial
### d) Gamma
### Evaluate which distribution fits which dataset best and justify probabilistically.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression, make_classification
from sklearn.linear_model import (
    LinearRegression,
    LogisticRegression,
    PoissonRegressor,
    GammaRegressor
)
from sklearn.metrics import mean_squared_error, accuracy_score

np.random.seed(42)
```

```python
# Gaussian GLM (Linear Regression)
X_gauss, y_gauss = make_regression(
    n_samples=200,
    n_features=1,
    noise=15,
    random_state=42
)

gaussian_glm = LinearRegression()
gaussian_glm.fit(X_gauss, y_gauss)

y_pred_gauss = gaussian_glm.predict(X_gauss)

print("GAUSSIAN GLM")
print("Coefficient:", gaussian_glm.coef_[0])
print("Intercept:", gaussian_glm.intercept_)
print("MSE:", mean_squared_error(y_gauss, y_pred_gauss))
```

```python
# Poisson GLM
X_pois = np.random.rand(200, 1)
y_pois = np.random.poisson(
    lam=np.exp(1 + 2 * X_pois.flatten())
)

poisson_glm = PoissonRegressor()
poisson_glm.fit(X_pois, y_pois)

y_pred_pois = poisson_glm.predict(X_pois)

print("\nPOISSON GLM")
print("Coefficient:", poisson_glm.coef_[0])
print("Intercept:", poisson_glm.intercept_)
print("Mean predicted count:", np.mean(y_pred_pois))
```

```python
# Binomial GLM (Logistic Regression)
X_bin, y_bin = make_classification(
    n_samples=300,
    n_features=2,
    n_redundant=0,
    random_state=42
)

binomial_glm = LogisticRegression(max_iter=2000)
binomial_glm.fit(X_bin, y_bin)

y_pred_bin = binomial_glm.predict(X_bin)
y_prob_bin = binomial_glm.predict_proba(X_bin)[:, 1]

print("\nBINOMIAL GLM")
print("Accuracy:", accuracy_score(y_bin, y_pred_bin))
print("Mean predicted probability:", np.mean(y_prob_bin))
```

```python
# Gamma GLM
X_gamma = np.random.rand(200, 1)
y_gamma = np.random.gamma(
    shape=2.0,
    scale=2.0,
    size=200
)

gamma_glm = GammaRegressor()
gamma_glm.fit(X_gamma, y_gamma)

y_pred_gamma = gamma_glm.predict(X_gamma)

print("\nGAMMA GLM")
print("Coefficient:", gamma_glm.coef_[0])
print("Intercept:", gamma_glm.intercept_)
print("Mean predicted value:", np.mean(y_pred_gamma))
```

### Generalized Linear Models extend linear regression by allowing the response variable to follow different exponential family distributions. Gaussian GLMs are appropriate for continuous data, Poisson GLMs for count data, Binomial GLMs for binary classification, and Gamma GLMs for positive skewed responses. The choice of distribution and link function should reflect the probabilistic nature of the data-generating process.
