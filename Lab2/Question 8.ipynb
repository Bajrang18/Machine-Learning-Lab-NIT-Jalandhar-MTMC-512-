### 8. Generate a non-linearly separable dataset (e.g., concentric circles or moons).
#### a) Train a Logistic Regression model.
#### b) Visualize decision boundaries.
#### c) Explain probabilistically why performance is limited.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

np.random.seed(42)

X, y = make_moons(
    n_samples=500,
    noise=0.25
)

plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm")
plt.title("Non-Linearly Separable Dataset (Moons)")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()

logreg = LogisticRegression()
logreg.fit(X, y)

y_pred = logreg.predict(X)
acc = accuracy_score(y, y_pred)

print("Logistic Regression Accuracy:", acc)


def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:,0].min()-0.5, X[:,0].max()+0.5
    y_min, y_max = X[:,1].min()-0.5, X[:,1].max()+0.5
    
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:,0], X[:,1], c=y, edgecolor="k")
    plt.title("Logistic Regression Decision Boundary")
    plt.show()

plot_decision_boundary(logreg, X, y)
```

Logistic regression performs poorly on non-linearly separable datasets because it assumes a linear relationship between the log-odds of the class probability and the input features. This results in linear decision boundaries that cannot capture complex class structures, leading to systematic misclassification despite correct probabilistic optimization.
