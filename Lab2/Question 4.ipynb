### 4. Implement linear regression from scratch using Maximum Likelihood Estimation (MLE).
#### a. Compare learned parameters with sklearn.linear_model.LinearRegression
#### b. Discuss numerical differences, if any.

```python
# Generate Synthetic Regression Dataset
import numpy as np
from sklearn.linear_model import LinearRegression

np.random.seed(42)

n = 200
X = np.random.randn(n, 1)
true_w = 3.5
true_b = 2.0

noise = np.random.normal(0, 1, n)
y = true_w * X.flatten() + true_b + noise

# Add Bias Term
X_design = np.c_[np.ones(n), X]   # [1, x]

w_mle = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ y

b_mle = w_mle[0]
w_mle_param = w_mle[1]

print("MLE Weight:", w_mle_param)
print("MLE Bias:", b_mle)

lr = LinearRegression()
lr.fit(X, y)

print("Sklearn Weight:", lr.coef_[0])
print("Sklearn Bias:", lr.intercept_)
```

### Linear regression under the assumption of Gaussian noise leads to a closed-form Maximum Likelihood Estimation solution identical to the least-squares estimator. The parameters learned via manual MLE implementation closely match those obtained using sklearn’s LinearRegression, with minor numerical differences arising from optimization techniques and floating-point precision.
