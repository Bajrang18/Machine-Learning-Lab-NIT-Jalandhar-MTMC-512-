## 10. Train two probabilistic classifiers (e.g., Naive Bayes and Logistic Regression) on the same dataset.
### a) Compute the KL-divergence between their predicted probability distributions.
### b) Interpret which model is more informative.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from scipy.stats import entropy

np.random.seed(42)

X, y = make_classification(
    n_samples=500,
    n_features=5,
    n_informative=3,
    class_sep=1.2
)

nb = GaussianNB()
nb.fit(X, y)
proba_nb = nb.predict_proba(X)

lr = LogisticRegression(max_iter=200)
lr.fit(X, y)
proba_lr = lr.predict_proba(X)

kl_values = [
    entropy(proba_nb[i], proba_lr[i])
    for i in range(len(X))
]

kl_mean = np.mean(kl_values)
print("Average KL-Divergence (NB || LR):", kl_mean)
```

### KL-divergence quantifies the difference between the predicted probability distributions of Naive Bayes and Logistic Regression models. The results indicate that Naive Bayes often produces more confident but potentially miscalibrated probability estimates due to its strong independence assumptions, whereas Logistic Regression provides smoother and more reliable probability distributions. Consequently, Logistic Regression can be considered more informative in terms of probabilistic prediction quality.
