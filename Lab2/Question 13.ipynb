### 13. Create a highly class-imbalanced dataset.
#### a) Train a probabilistic classifier.
#### b) c) Analyze how probability estimates become misleading.
#### c) Apply Bayesian reweighting or class priors to correct this.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report


np.random.seed(42)

X, y = make_classification(
    n_samples=5000,
    n_features=5,
    n_informative=3,
    n_redundant=0,
    weights=[0.95, 0.05],   # 95% vs 5%
    flip_y=0.01
)

print("Class distribution:", np.bincount(y))


```

```python
lr = LogisticRegression(max_iter=500)
lr.fit(X, y)

proba = lr.predict_proba(X)[:, 1]
y_pred = lr.predict(X)

print(classification_report(y, y_pred))
```

```python
print("Average predicted probability for positive class:",
      np.mean(proba[y == 1]))
```

```python
lr_balanced = LogisticRegression(
    class_weight={0:1, 1:19},  # inverse of class ratio
    max_iter=500
)
lr_balanced.fit(X, y)

proba_bal = lr_balanced.predict_proba(X)[:, 1]
y_pred_bal = lr_balanced.predict(X)

print(classification_report(y, y_pred_bal))
```

In highly imbalanced datasets, probabilistic classifiers tend to produce misleadingly low posterior probabilities for the minority class due to dominant class priors. Although this behavior is consistent with Bayes’ theorem, it leads to poor detection of rare but important events. By applying Bayesian reweighting through adjusted class priors or class-weighted training, the posterior probabilities become more informative and suitable for decision-making under imbalance.
