## 19. Use a synthetic regression or classification dataset generated using sklearn.datasets.make_regression or make_classification (choose one).
### a. Fit two competing probabilistic models (e.g., linear vs polynomial regression, or Logistic Regression vs Naive Bayes).
### b. Compute the log-likelihood of each model on the same dataset.
### c. Perform a Likelihood Ratio Test (LRT) to statistically compare the models.
### d. Interpret the test statistic and decision at a chosen significance level.

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from scipy.stats import chi2

np.random.seed(42)

n = 200
X = np.linspace(-3, 3, n).reshape(-1, 1)

# True quadratic relationship
y = 1.5 * X.flatten()**2 + 2 * X.flatten() + 1
y += np.random.normal(0, 1.0, n)

lin_model = LinearRegression()
lin_model.fit(X, y)

y_pred_lin = lin_model.predict(X)
residuals_lin = y - y_pred_lin

poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)

y_pred_poly = poly_model.predict(X_poly)
residuals_poly = y - y_pred_poly



sigma2_lin = np.var(residuals_lin)
sigma2_poly = np.var(residuals_poly)



def log_likelihood(residuals, sigma2):
    n = len(residuals)
    return -0.5 * n * np.log(2 * np.pi * sigma2) \
           - 0.5 * np.sum(residuals**2) / sigma2

LL_lin = log_likelihood(residuals_lin, sigma2_lin)
LL_poly = log_likelihood(residuals_poly, sigma2_poly)

print("Log-Likelihood (Linear):", LL_lin)
print("Log-Likelihood (Polynomial):", LL_poly)


LR_stat = 2 * (LL_poly - LL_lin)
print("Likelihood Ratio Statistic:", LR_stat)


p_value = 1 - chi2.cdf(LR_stat, df=1)
print("p-value:", p_value)
```

## Two nested probabilistic regression models were compared using a Likelihood Ratio Test. The polynomial regression model achieved a significantly higher log-likelihood than the linear model. The resulting likelihood ratio statistic exceeded the critical value at the 5% significance level, leading to rejection of the null hypothesis. This indicates that the additional quadratic term provides a statistically significant improvement in explaining the data.
