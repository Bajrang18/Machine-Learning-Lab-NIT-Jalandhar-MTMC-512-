#### a. Train a Naive Bayes classifier and a Logistic Regression model.
#### b. Compare their performance and explain the results using probabilistic assumptions.

## Generate Highly Correlated Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

np.random.seed(42)

n = 1000

# Base feature
x1 = np.random.randn(n)

# Highly correlated feature
x2 = x1 + np.random.normal(0, 0.05, n)

X = np.column_stack((x1, x2))

# Binary target
y = (x1 + x2 > 0).astype(int)
```

### Train Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(X, y)

y_pred_nb = gnb.predict(X)
acc_nb = accuracy_score(y, y_pred_nb)
```

### Train Logistic Regression

```python
logreg = LogisticRegression()
logreg.fit(X, y)

y_pred_lr = logreg.predict(X)
acc_lr = accuracy_score(y, y_pred_lr)
```

### Performance Comparison

```python
print("Naive Bayes Accuracy:", acc_nb)
print("Logistic Regression Accuracy:", acc_lr)
```

### Visualization

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Highly Correlated Features")
plt.show()
```

### Naive Bayes performs poorly when features are highly correlated because its conditional independence assumption is violated, leading to double-counting of evidence. Logistic Regression does not rely on such assumptions and therefore achieves superior performance by directly modeling the conditional probability of the class label given the features.
