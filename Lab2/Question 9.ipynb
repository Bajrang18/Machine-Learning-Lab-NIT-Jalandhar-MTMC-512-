## 9. For a 2D linear regression problem:
### a) Compute and plot the likelihood surface over two parameters.
### b) Identify regions of high and low likelihood.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 50
X = np.linspace(-3, 3, n)
true_w = 2.0
true_b = 1.0

noise = np.random.normal(0, 0.5, n)
y = true_w * X + true_b + noise


def log_likelihood(w, b, X, y, sigma2=0.25):
    y_pred = w * X + b
    return -0.5 * np.sum(
        np.log(2 * np.pi * sigma2) +
        (y - y_pred)**2 / sigma2
    )



w_vals = np.linspace(0, 4, 100)
b_vals = np.linspace(-1, 3, 100)

W, B = np.meshgrid(w_vals, b_vals)
LL = np.zeros_like(W)

for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        LL[i, j] = log_likelihood(W[i, j], B[i, j], X, y)



plt.contourf(W, B, LL, levels=50, cmap="viridis")
plt.colorbar(label="Log-Likelihood")
plt.xlabel("Weight (w)")
plt.ylabel("Bias (b)")
plt.title("Log-Likelihood Surface")
plt.scatter(true_w, true_b, color="red", label="True Parameters")
plt.legend()
plt.show()
```

The likelihood surface for a 2D linear regression model is smooth and convex with a single global maximum corresponding to the Maximum Likelihood Estimate of the parameters. Regions near the peak represent parameter values with high explanatory power, while regions far away indicate poor model fit. This visualization provides geometric intuition for parameter estimation and optimization in probabilistic models.
