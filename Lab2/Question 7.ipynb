### 7. Use a binary classification dataset (make_classification).
#### a. Train logistic regression models using different link functions (logit, probit).
#### b. Compare convergence behavior and classification performance.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from scipy.stats import norm

np.random.seed(42)

X, y = make_classification(
    n_samples=500,
    n_features=5,
    n_informative=3,
    n_redundant=0,
    class_sep=1.2
)

logit_model = LogisticRegression(
    solver="lbfgs",
    max_iter=200
)
logit_model.fit(X, y)

y_pred_logit = logit_model.predict(X)
acc_logit = accuracy_score(y, y_pred_logit)

print("Logit Accuracy:", acc_logit)
print("Logit Iterations:", logit_model.n_iter_)


import statsmodels.api as sm

X_sm = sm.add_constant(X)

probit_model = sm.Probit(y, X_sm)
probit_result = probit_model.fit(disp=False)

y_pred_probit = (probit_result.predict(X_sm) > 0.5).astype(int)
acc_probit = accuracy_score(y, y_pred_probit)

print("Probit Accuracy:", acc_probit)
print("Probit Iterations:", probit_result.mle_retvals['iterations'])
```

Logistic regression models using logit and probit link functions show comparable classification performance due to the similar sigmoidal shapes of their response functions. However, the logit model converges faster because of its closed-form gradient and simpler likelihood structure, while the probit model, though probabilistically well-motivated, requires numerical optimization and converges more slowly.
