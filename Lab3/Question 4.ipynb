## 4. Create a synthetic classification dataset using make_classification. Train two Decision Tree models: one using criterion='gini' and the other using criterion='entropy'.
* ## Task: Visualize both trees using plot_tree. Identify one node where the split differs and explain the mathematical difference between Gini Impurity and Information Gain (Entropy).
* ## Task: Visualize the first two levels of each tree. Do they pick the same feature for the root split?

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Create synthetic data
X, y = make_classification(
    n_samples=200,
    n_features=4,
    n_informative=2,
    random_state=42
)
```

```python
tree_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
tree_entropy = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)

tree_gini.fit(X, y)
tree_entropy.fit(X, y)
```

```python
plt.figure(figsize=(12,5))
plot_tree(tree_gini, filled=True)
plt.title("Decision Tree using Gini")
plt.show()

plt.figure(figsize=(12,5))
plot_tree(tree_entropy, filled=True)
plt.title("Decision Tree using Entropy")
plt.show()
```

### The tree built using Gini impurity and the one built using entropy both aim to create pure nodes but use different mathematical measures. Gini impurity calculates how often a randomly chosen element would be misclassified, whereas entropy measures the amount of uncertainty or information content. In this experiment, the first split chosen by both trees was often the same because one feature provided strong separation. However, deeper nodes sometimes differed due to the different sensitivity of the impurity measures.
