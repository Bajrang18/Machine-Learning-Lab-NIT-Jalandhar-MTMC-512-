# 3. The choice of "distance" defines what the model considers "close." Using the Wine dataset:
* # Task: Compare metric='euclidean' (L2 norm) and metric='manhattan' (L1 norm).
* # Calculation: Explain how the model calculates the distance for a single test point in both cases using the formulas:
 * ## Euclidean: 𝑑(𝑥, 𝑦) = √{∑ (𝑥𝑖− 𝑦𝑖)2}
 * ## Manhattan: 𝑑(𝑥, 𝑦) = ∑ |𝑥𝑖− 𝑦𝑖|

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load dataset
wine = load_wine()
X, y = wine.data, wine.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, #random_state=42
)
```

```python
# Scale data (important for k-NN)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
# Euclidean Distance (L2)
knn_e = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_e.fit(X_train_s, y_train)
pred_e = knn_e.predict(X_test_s)
acc_e = accuracy_score(y_test, pred_e)
```

```python
# Manhattan Distance (L1)
knn_m = KNeighborsClassifier(n_neighbors=5, metric='manhattan')
knn_m.fit(X_train_s, y_train)
pred_m = knn_m.predict(X_test_s)
acc_m = accuracy_score(y_test, pred_m)
```

```python
print("Euclidean Accuracy:", acc_e)
print("Manhattan Accuracy:", acc_m)
```

#### The Euclidean metric computes the straight-line distance and emphasizes large feature differences due to squaring. Manhattan distance sums absolute differences and is less sensitive to extreme values. In the Wine dataset, both metrics produce comparable accuracy, but Manhattan can perform slightly better when outliers are present. The choice of metric affects the shape of the decision boundary and the way neighbors are selected.
