## 6. Decision Trees are notorious for overfitting. Using the Breast Cancer dataset, train a tree with no depth limit. Then, train another tree with max_depth=3 and min_samples_leaf=5.
* ## Task: Compare the training accuracy vs. testing accuracy for both. Explain the "bias- variance tradeoff" in the context of tree depth.

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

wine = load_wine()
X, y = wine.data, wine.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
tree_shallow = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_shallow.fit(X_train, y_train)

train_pred_shallow = tree_shallow.predict(X_train)
test_pred_shallow = tree_shallow.predict(X_test)

print("Shallow Tree Training Accuracy:",
      accuracy_score(y_train, train_pred_shallow))

print("Shallow Tree Testing Accuracy:",
      accuracy_score(y_test, test_pred_shallow))
```

```python
tree_deep = DecisionTreeClassifier(max_depth=None, random_state=42)
tree_deep.fit(X_train, y_train)

train_pred_deep = tree_deep.predict(X_train)
test_pred_deep = tree_deep.predict(X_test)

print("Deep Tree Training Accuracy:",
      accuracy_score(y_train, train_pred_deep))

print("Deep Tree Testing Accuracy:",
      accuracy_score(y_test, test_pred_deep))
```

## The shallow tree with max_depth=2 shows lower training accuracy because it cannot capture complex patterns, indicating higher bias but better generalization. The fully grown tree achieves near-perfect training accuracy but performs worse on test data, showing high variance and overfitting. This experiment demonstrates the bias–variance tradeoff, where increasing model complexity reduces bias but increases variance.
