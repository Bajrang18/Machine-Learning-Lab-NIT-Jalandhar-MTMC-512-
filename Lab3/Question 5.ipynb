# 5. Decision trees can handle both "Yes/No" data and numbers.
* ## Task: Create a small dummy DataFrame (e.g., Features: 'Is it Raining?', 'Temperature'; Target: 'Go to Park?').
* ## Task: Use 𝐿𝑎𝑏𝑒𝑙𝐸𝑛𝑐𝑜𝑑𝑒𝑟 to turn the "Yes/No" into 0s and 1s, then train a tree. Explain how the tree treats a "1" vs a temperature of "90°F."

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Create dataset
data = {
    "Is_it_Raining": ["Yes","No","No","Yes","No","Yes","No","No"],
    "Temperature": [65,85,90,70,95,60,88,92],
    "Go_to_Park": ["No","Yes","Yes","No","Yes","No","Yes","Yes"]
}

df = pd.DataFrame(data)

# Encode Yes/No
le = LabelEncoder()
df["Is_it_Raining"] = le.fit_transform(df["Is_it_Raining"])
df["Go_to_Park"] = le.fit_transform(df["Go_to_Park"])

# Features & target
X = df[["Is_it_Raining","Temperature"]]
y = df["Go_to_Park"]

# Train model
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# Plot tree
plt.figure(figsize=(10,5))
plot_tree(model, feature_names=X.columns, filled=True)
plt.show()
```

## LabelEncoder converts Yes/No into numeric values, but the decision tree still treats them as categorical splits because the thresholds separate 0 and 1 values. In contrast, temperature is treated as a continuous feature where the tree finds optimal numeric cut points. This shows that decision trees naturally handle mixed feature types without requiring feature scaling.
