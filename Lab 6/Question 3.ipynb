Question 3: Address the problem of "overfitting" in GMMs by determining the optimal number of components K.
Task:
1) Load a multidimensional dataset (e.g., Iris or Wine dataset).
2) Fit GMMs with the number of components K ranging from 1 to 10.
3) Compute and plot the Akaike Information Criterion (AIC) and the Bayesian Information Criterion (BIC) for each K.
4) Identify the "elbow" or minimum point. If AIC and BIC suggest different values for K, explain the theoretical reason behind this discrepancy.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

from sklearn.datasets import load_wine
from sklearn.mixture import GaussianMixture

data = load_wine()
X = data.data

aic = []
bic = []
K_range = range(1, 11)

for k in K_range:
    gmm = GaussianMixture(n_components=k)
    gmm.fit(X)

    aic.append(gmm.aic(X))
    bic.append(gmm.bic(X))

plt.plot(K_range, aic, label="AIC")
plt.plot(K_range, bic, label="BIC")
plt.legend()
plt.show()
```
