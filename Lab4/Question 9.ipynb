### 9. Implement a simple 3-layer Neural Network (Input, Hidden, Output) using only NumPy. You must manually code:
### a) The forward pass.
### b) The loss function (MSE or Cross-Entropy).
### c) The backpropagation algorithm using the chain rule to update weights.

```python
import numpy as np

# XOR Dataset
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0],[1],[1],[0]])

# Initialize Weights
np.random.seed(42)

input_size = 2
hidden_size = 4
output_size = 1

W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

learning_rate = 0.1
epochs = 5000

# Activation Functions
def relu(Z):
    return np.maximum(0, Z)

def relu_derivative(Z):
    return Z > 0

def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

# Training Loop
for epoch in range(epochs):

    # Forward pass
    Z1 = X @ W1 + b1
    A1 = relu(Z1)

    Z2 = A1 @ W2 + b2
    A2 = sigmoid(Z2)

    # Compute loss (binary cross entropy)
    loss = -np.mean(y*np.log(A2+1e-8) + (1-y)*np.log(1-A2+1e-8))

    # Backpropagation
    dZ2 = A2 - y
    dW2 = A1.T @ dZ2
    db2 = np.sum(dZ2, axis=0, keepdims=True)

    dA1 = dZ2 @ W2.T
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = X.T @ dZ1
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # Update weights
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1

    if epoch % 1000 == 0:
        print("Epoch:", epoch, "Loss:", loss)

# Final Predictions
Z1 = X @ W1 + b1
A1 = relu(Z1)
Z2 = A1 @ W2 + b2
A2 = sigmoid(Z2)

predictions = (A2 > 0.5).astype(int)

print("\nFinal Predictions:")
print(predictions)
```

### A simple neural network with one hidden layer was implemented using NumPy. The network successfully learned the XOR pattern through forward propagation and backpropagation using gradient descent. This demonstrates that multi-layer neural networks can model nonlinear relationships, unlike single-layer linear models.
