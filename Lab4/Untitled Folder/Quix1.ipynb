Gaussian Naive Bayes (GNB) and Logistic Regression are both commonly used classification algorithms. While Logistic Regression is often more powerful on large datasets, Gaussian Naive Bayes can sometimes outperform it when the dataset is very small.

This report explains why this happens.

Why GNB Performs Better on Very Small Datasets
 Fewer Parameters to Estimate

GNB estimates mean and variance for each feature per class.

Logistic Regression estimates one weight per feature plus bias.

When dataset is very small:

Logistic Regression struggles to estimate stable weights.

GNB remains stable because it relies on strong assumptions.

Strong Bias Helps in Small Data

GNB has strong assumptions (feature independence).
This introduces high bias but low variance.

In small datasets:

Low variance is more important.

Logistic Regression may overfit due to higher variance.


```python

import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


wine = load_wine()
X = wine.data      # continuous features
y = wine.target


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred_gnb = gnb.predict(X_test)


log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_scaled, y_train)

y_pred_log = log_reg.predict(X_test_scaled)


print("Gaussian Naive Bayes Accuracy:",
      accuracy_score(y_test, y_pred_gnb))

print("Logistic Regression Accuracy:",
      accuracy_score(y_test, y_pred_log))

print("\nClassification Report (Logistic Regression):")
print(classification_report(y_test, y_pred_log))
```

Logistic regression models using logit and probit link functions show comparable classification performance due to the similar sigmoidal shapes of their response functions. However, the logit model converges faster because of its closed-form gradient and simpler likelihood structure, while the probit model, though probabilistically well-motivated, requires numerical optimization and converges more slowly.
