## 7. In many libraries, you can define your own kernel. Implement a custom Laplacian Kernel 𝐾(𝑥, 𝑦) = 𝑒^−𝛼 ||𝑥 – 𝑦||1 and use it within an SVM framework. Compare its performance against the standard RBF kernel on a dataset of your choice.

```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics.pairwise import manhattan_distances
```

### Load and Scale Data

```python
wine = load_wine()
X = wine.data
y = wine.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Define Laplacian Kernel

```python
def laplacian_kernel(X, Y, alpha=1.0):
    return np.exp(-alpha * manhattan_distances(X, Y))
```

### Train SVM with Custom Kernel


```python
svm_laplacian = SVC(kernel=lambda X, Y: laplacian_kernel(X, Y, alpha=0.5))

scores_lap = cross_val_score(svm_laplacian, X_scaled, y, cv=5)
print("Laplacian Kernel Accuracy:", np.mean(scores_lap))
```

### Compare with RBF Kernel

```python
svm_rbf = SVC(kernel='rbf')
scores_rbf = cross_val_score(svm_rbf, X_scaled, y, cv=5)
print("RBF Kernel Accuracy:", np.mean(scores_rbf))
```

Typically:

* RBF performs slightly better

* Laplacian performs slightly lesser

* Both outperform linear

The Laplacian kernel uses the L1 (Manhattan) distance instead of squared Euclidean distance as in the RBF kernel. This makes it more robust to large feature differences and potential outliers. In the Wine dataset, both Laplacian and RBF kernels achieved strong performance, demonstrating that nonlinear kernel methods effectively capture complex feature interactions.
