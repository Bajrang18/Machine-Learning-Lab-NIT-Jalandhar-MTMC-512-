## 5. Using the UCI Wine dataset, compare the classification accuracy of an SVM using three different kernels: Linear, Polynomial (degree 3), and RBF. Which one performs best? Hypothesize why based on the feature distribution.

```python
import numpy as np
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

### Load Dataset

```python
wine = load_wine()
X = wine.data
y = wine.target
```

### Scale Features

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Define Kernels

```python
kernels = {
    "Linear": SVC(kernel='linear'),
    "Polynomial": SVC(kernel='poly', degree=3),
    "RBF": SVC(kernel='rbf')
}
```

### Cross Validation

```python
results = {}

for name, model in kernels.items():
    scores = cross_val_score(model, X_scaled, y, cv=5)
    results[name] = np.mean(scores)
```

### print Results

```python
for kernel, score in results.items():
    print(f"{kernel} Kernel Accuracy: {score:.4f}")
```

### The linear kernel assumes a straight decision boundary and performs well when the data is linearly separable. The polynomial kernel introduces nonlinearity by modeling feature interactions. The RBF kernel uses a Gaussian similarity function, allowing highly flexible decision boundaries. In the Wine dataset, the RBF kernel achieved the highest cross-validation accuracy because it effectively captured the nonlinear relationships among feteures
