### 11. Experiment with weight initialization. Compare the training progress of a deep network (5+ layers) when weights are initialized to all zeros and large random numbers. Observe and document the "vanishing gradient" problem.

## Small Demonstration Code

### Zero Initialization Example

```python
import numpy as np

X = np.random.randn(5,3)
W = np.zeros((3,4))

Z = X @ W
print("Output with zero weights:\n", Z)
```

### All outputs identical → no learning diversity.

## Large Initialization Example

```python
W_large = np.random.randn(3,4) * 100

Z_large = X @ W_large
print("Large activations:\n", Z_large)
```

### Values become extremely large → sigmoid saturates.

## nitializing all weights to zero causes symmetry, preventing neurons from learning different features. Very large initial weights cause saturation in sigmoid or tanh activations, leading to vanishing gradients. Proper initialization methods such as Xavier and He initialization maintain stable gradient flow and improve training stability in deep neural networks.
