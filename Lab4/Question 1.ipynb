### 1. Implement the closed-form solution for Ridge Regression: = (𝑋𝑇𝑋 + 𝜆 𝐼)−1 𝑋𝑇𝑦. Using a dataset with high multicollinearity (like the Boston Housing or Diabetes dataset), plot how the magnitude of the coefficients (𝛽 weights) changes as 𝜆 increases from 0 to 104.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler

# Load dataset
data = load_diabetes()
X = data.data
y = data.target

# Scale features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Closed-form Ridge
def ridge_closed_form(X, y, lam):
    n_features = X.shape[1]
    I = np.eye(n_features)
    return np.linalg.inv(X.T @ X + lam * I) @ X.T @ y

lambdas = np.logspace(0, 4, 50)
coefficients = []

for lam in lambdas:
    beta = ridge_closed_form(X, y, lam)
    coefficients.append(beta)

coefficients = np.array(coefficients)

# Plot
plt.figure(figsize=(8,6))
for i in range(coefficients.shape[1]):
    plt.plot(lambdas, coefficients[:, i])

plt.xscale('log')
plt.xlabel("Lambda")
plt.ylabel("Coefficient Value")
plt.title("Ridge Coefficient Shrinkage")
plt.show()
```

Explanation

As λ increases:

* Coefficients shrink toward zero

* Model becomes more stable

* Multicollinearity impact reduces.
