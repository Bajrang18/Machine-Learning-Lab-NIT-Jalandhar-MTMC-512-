## 8.  A single-layer Perceptron cannot solve the Exclusive-OR (XOR) problem.
* ### Prove this by attempting to train a Perceptron on XOR data.
* ### Build a Multi-Layer Perceptron (MLP) with one hidden layer (2-4 neurons) that successfully classifies XOR. Plot the decision boundary.

```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.neural_network import MLPClassifier
```

### Create XOR Dataset 

```python
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,0])
```

## Train Perceptron

```python
perceptron = Perceptron(max_iter=1000)
perceptron.fit(X, y)

print("Perceptron Accuracy:", perceptron.score(X,y))
```

## Train MLP

```python
mlp = MLPClassifier(hidden_layer_sizes=(4,),
                    activation='relu',
                    max_iter=5000)

mlp.fit(X, y)

print("MLP Accuracy:", mlp.score(X,y))
```

MLP has hidden layer.

Hidden layer allows model to create nonlinear transformation.

It can create intermediate representations:

Example:

Hidden neuron 1 → detects x1 AND NOT x2
Hidden neuron 2 → detects NOT x1 AND x2

Final output → OR of those two

That implements XOR.

The XOR dataset is not linearly separable, meaning no straight line can separate the two classes. A single-layer perceptron can only learn linear decision boundaries, so it fails to correctly classify XOR. However, a multi-layer perceptron introduces hidden layers and nonlinear activation functions, enabling it to model nonlinear relationships and perfectly classify XOR.
