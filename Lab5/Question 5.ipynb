```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)
```

```python
score = silhouette_score(X, labels)
print("Silhouette Score:", score)
```

```python
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    print("k:", k, "Score:", score)
```

## The silhouette score measures how similar a data point is to its own cluster compared to other clusters. A high silhouette value indicates well-separated clusters, while a negative value suggests misclassification. It helps evaluate cluster quality and select the optimal number of clusters.
