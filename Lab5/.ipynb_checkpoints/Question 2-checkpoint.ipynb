## 2. Standard K-Means initializes centroids randomly, which can lead to poor local minima or "empty clusters." K-Means++ was designed to fix this.
* ## Task: Run a standard K-Means and a K-Means++ implementation 50 iteration each on a dataset with highly dispersed clusters.
* ## The Question: Compare the distribution of the final inertia values and the number of iterations to converge. How does the probabilistic initialization strategy of K- Means++ specifically help in avoiding the "initialization trap" compared to purely random selection?

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import numpy as np

# from sklearn.datasets import load_wine

# data = load_wine()
# X = data.data


# Create dataset
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

inertia_random = []
inertia_plus = []

for _ in range(50):
    km_random = KMeans(n_clusters=3, init='random', n_init=1)
    km_plus = KMeans(n_clusters=3, init='k-means++', n_init=1)

    km_random.fit(X)
    km_plus.fit(X)

    inertia_random.append(km_random.inertia_)
    inertia_plus.append(km_plus.inertia_)

print("Average Inertia (Random):", np.mean(inertia_random))
print("Average Inertia (K-Means++):", np.mean(inertia_plus))
```

## K-Means++ improves clustering by selecting initial centroids that are far apart from each other. This reduces the chances of poor initialization and helps avoid local minima. As a result, K-Means++ typically achieves lower inertia and more stable clustering compared to random initialization.
