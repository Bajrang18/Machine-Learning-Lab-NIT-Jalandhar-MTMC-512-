## 3: DBSCAN doesn't care about 𝑘; it cares about density. The two primary hyperparameters are 𝜖 and 𝑚𝑖𝑛_𝑠𝑎𝑚𝑝𝑙𝑒𝑠.
* ## Task: Use a dataset with varying densities (like the "Moons" or "Concentric Circles" dataset). Set min_𝑠𝑎𝑚𝑝𝑙𝑒𝑠 = 5 and vary 𝜖 from a very small value to a very large value.
* ## The Question: Describe what happens to the number of clusters and the number of noise points (outliers) at both extremes of 𝜖. If your dataset has clusters of significantly different densities, why might DBSCAN struggle to find all of them with a single 𝜖 value?

```python
# QUESTION 3 — DBSCAN (Varying ε)
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

X, _ = make_moons(n_samples=300, noise=0.1)

eps_values = [0.05, 0.1, 0.2, 0.5, 1.0]

for eps in eps_values:
    db = DBSCAN(eps=eps, min_samples=5)
    labels = db.fit_predict(X)
    print("Eps:", eps,
          "Clusters:", len(set(labels)) - (1 if -1 in labels else 0),
          "Noise points:", list(labels).count(-1))
```

Small ε:

* Many noise points

* Many tiny clusters

Large ε:

* Few clusters

* Almost no noise

Why DBSCAN struggles with varying densities?
Because one ε cannot capture both dense and sparse clusters simultaneously.
